**Question-1 : Determine the Number of Tags Per Question**

```python
count=[]
unique_tag = dict()
for i in Ds3['tags']:
  #print(i)
  count.append(len(i.split('>'))-1)
  tg=i.split('>')
  tg.pop()
  for j in tg:
    if (j[1:] in unique_tag):
      unique_tag[j[1:]] += 1
    else:
      unique_tag[j[1:]] = 1
print("No. of tags per question:",count)
```
